```python
import pandas as pd
```

```python
df_olympic_swimming = pd.read_csv ('/Olympic_Swimming_Results_1912to2020.csv')
df_olympic_swimming
```

```python
df_filt_swimming = df_olympic_swimming
```

```python
# Foco no ano de 2016.

df_filt_swimming = df_olympic_swimming[(df_olympic_swimming['Year'] == 2016) & \
                                       ((df_olympic_swimming['Location'] == '') | \
                                        (df_olympic_swimming['Location'].isnull())) & \
                                       ((df_olympic_swimming['Gender'] == 'Men') | \
                                        (df_olympic_swimming['Gender'] == 'Women'))]


df_olympic_swimming
```

```python
# Verificar valores nulos em cada coluna do DataFrame df_filt_swimming
null_counts = df_filt_swimming.isnull().sum()

# Exibir a contagem de valores nulos por coluna
print("Contagem de valores nulos por coluna:")
print(null_counts)

# Filtrar linhas com valores nulos em qualquer coluna
df_filt_swimming_nulls = df_filt_swimming[df_filt_swimming.isnull().any(axis=1)]

# Exibir as primeiras 10 linhas do DataFrame com valores nulos
print("\nPrimeiras 10 linhas com valores nulos:")
print(df_filt_swimming_nulls.head(10))
```

```python
# Salvando os dez países com maior número de participações
df_olympic_swimming_superior = df_filt_swimming.sort_values('Location', ascending=False)
df_olympic_swimming_superior.head(3)
```

```python
df_olympic_swimming_inferior = df_filt_swimming.sort_values('Athlete', ascending=False)
df_olympic_swimming_inferior = df_olympic_swimming.head(3)
```
